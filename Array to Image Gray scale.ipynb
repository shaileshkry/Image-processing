```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
arr=np.zeros((8,12),dtype=int)
```

```python
len(arr)
```

```python
len(arr[0])
```

```python
for i in range(len(arr)):
    for j in range(len(arr[0])):
        if (i+j)%2==0:
            
            arr[i][j]=1
print(arr)
```

```python
plt.imshow(arr)
```

```python
cv.imwrite('chess_board.png',arr)
```

```python
for i in range(len(arr)):
    for j in range(len(arr[0])):
        if (i+j)%2==0:
            
            arr[i][j]=250-(3*i*j)
print(arr)
```

```python
plt.imshow(arr)
```

```python
cv.imwrite('chess_board2.png',arr)
```

```python
for i in range(len(arr)):
    for j in range(len(arr[0])):
        if (i+j)%2!=0:
            
            arr[i][j]=250-(3*i*j)
print(arr)
```

```python
plt.imshow(arr)
```

```python
arr=np.zeros((720,1080),dtype=int)
for i in range(len(arr)):
    for j in range(len(arr[0])):
        if (i+j)%2!=0:
            
            arr[i][j]=(i+j)%250
print(arr)
```

```python
plt.imshow(arr)
```

```python
cv.imwrite('Frame1.png',arr)
```

```python
arr=np.zeros((720,1080),dtype=int)
for i in range(len(arr)):
    for j in range(len(arr[0])):
        if (i+j)%2!=0:
            
            arr[i][j]=(i*j)%250
print(arr)
```

```python
plt.imshow(arr)
```

```python
cv.imwrite('Frame2.png',arr)
```
