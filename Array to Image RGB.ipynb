```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
arr=np.zeros((3,3),dtype=int)
```

```python
len(arr)
```

```python
len(arr[0])
```

```python
for i in range(len(arr)):
    for j in range(len(arr[0])):
        if (i+j)%2==0:
            
            arr[i][j]=1
print(arr)
```

```python
plt.imshow(arr)
```

### RGB

```python
lst=[]
for i in range(len(arr)):
    lst1=[]
    for j in range(len(arr[0])):
        lst2=[]
        for k in range(0,3):
            lst2.append(250-(30*k*i*j))
        lst1.append(lst2)
    lst.append(lst1)
print(lst)
arr=np.array(lst)
```

```python
plt.imshow(arr)
```

```python
cv.imwrite('chess_board2.png',arr)
```

```python
lst=[]
for i in range(720):
    lst1=[]
    for j in range(1080):
        lst2=[]
        for k in range(0,3):
            lst2.append((i+j+k)%250)
        lst1.append(lst2)
    lst.append(lst1)
print(lst)
arr=np.array(lst)
```

```python
lst=[]
for i in range(720):
    lst1=[]
    for j in range(1080):
        lst2=[]
        for k in range(0,3):
            if (i+j)%5==0:
                if k==0:
                    lst2.append((i+j+k)%250)
                else:
                    lst2.append(0)
                    
            else:
                lst2.append(0)
        lst1.append(lst2)
    lst.append(lst1)
print(lst)
arr=np.array(lst)
```

```python
plt.imshow(arr)
```

```python
cv.imwrite('Frame2col.png',arr)
```

```python
lst=[]
for i in range(720):
    lst1=[]
    for j in range(1080):
        lst2=[]
        for k in range(0,3):
            if (i+j)%5==0:
                if k==0:
                    lst2.append((i+j+k)%250)
                elif k==1:
                    lst2.append(0)
                else:
                    lst2.append(250)
                    
            else:
                lst2.append(0)
        lst1.append(lst2)
    lst.append(lst1)
print(lst)
arr=np.array(lst)
plt.imshow(arr)
cv.imwrite('Frame3col.png',arr)
```
