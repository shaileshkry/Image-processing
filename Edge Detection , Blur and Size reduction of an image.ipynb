```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img=cv.imread('ram.jpg')
```

```python
plt.imshow(img)
```

```python
#Edge Detection
plt.imshow(cv.Canny(img,100,610))
```

```python
img=cv.cvtColor(img,cv.COLOR_RGB2BGR)
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv.Canny(img,100,610))
```

```python
#Blur image
plt.imshow(cv.blur(img,(2,2)))
```

```python
plt.imshow(cv.blur(img,(4,4)))
```

```python
plt.imshow(cv.blur(img,(8,8)))
```
